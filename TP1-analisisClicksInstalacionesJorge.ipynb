# Analizo installs

```python
inst = pd.read_csv("data/installs.csv", dtype = {"application_id": np.int16, "click_hash": np.float16})
```

```python
inst.memory_usage(deep=True).sum()
```

```python
del inst['event_uuid']
del inst['click_hash']
del inst['trans_id']
inst['created']=pd.to_datetime(inst['created'])
inst['wifi']=inst['wifi'].astype('bool')
inst['session_user_agent']=inst['session_user_agent'].dropna()
inst['session_user_agent']=inst['session_user_agent'].astype('category')

inst['user_agent']=inst['user_agent'].astype('category')
inst['device_countrycode']=inst['device_countrycode'].astype('category')
inst['device_brand']=inst['device_brand'].astype('category')
inst['device_language']=inst['device_language'].astype('category')
inst['kind']=inst['kind'].astype('category')
```

```python
#funcion que recibe como argumento un dia en ingles y lodevuelve el dia en español
def diaEspañol(dia):
    if (dia=="Monday"): return "Lunes"
    if (dia=="Tuesday"): return "Martes"
    if (dia=="Wednesday"): return "Miercoles"
    if (dia=="Thursday"): return "Jueves"
    if (dia=="Friday"): return "Viernes"
    if (dia=="Saturday"): return "Sabado"
    if (dia=="Sunday"): return "Domingo"
    else: return dia
```

```python
inst.memory_usage(deep=True).sum()
```

```python
inst['año'] = pd.DatetimeIndex(inst['created']).year
inst['mes'] =inst['created'].dt.month
inst['día'] = inst['created'].dt.day
inst['hora'] = inst['created'].dt.hour
inst['minuto'] = inst['created'].dt.minute
inst['diaSemana'] = inst['created'].dt.weekday_name
inst.head()
```

```python
inst['diaSemana']=inst['diaSemana'].apply(diaEspañol)
```

```python
grupo_inst_por_hora=inst.groupby(['diaSemana','hora'])['hora'].count().sort_values(ascending=False).rename("Cantidad de instalaciones").reset_index()
grupo_inst_por_hora.head()
```

```python
df2 = grupo_inst_por_hora[['diaSemana', 'hora','Cantidad de instalaciones']]
heatmap2_data = pd.pivot_table(df2,values='Cantidad de instalaciones', index=['diaSemana'], columns={'hora'})
heatmap2_data.head(n=5)
ax=sns.heatmap(heatmap2_data, cmap="BuGn")
ax.set_title('Instalaciones totales por  día y hora',fontsize = 25)
ax.set_xlabel("Hora", fontsize = 15)
ax.set_ylabel("Día", fontsize = 15)
ax.figure.set_size_inches(16, 7)
figure = ax.get_figure()   
figure.savefig("Instalaciones_por_dia_hora.png")
```

```python
grupo_inst=inst.groupby(['application_id'])['application_id'].count().sort_values(ascending=False).rename("Cantidad").reset_index()

grupo_inst.head()
```

```python
top10_app=grupo_inst['application_id'].head(10)
top10_app
```

```python
grupo_inst_por_app=inst.groupby(['application_id','diaSemana'])['application_id'].count().sort_values(ascending=False).rename("Total instalaciones").reset_index()
grupo_inst_por_app['application_id']=grupo_inst_por_app[grupo_inst_por_app['application_id'].isin(top10_app)]
grupo_inst_por_app.head(20)
```

```python
ax = sns.swarmplot(x="Total instalaciones", y="diaSemana", hue="application_id", size=18,data=grupo_inst_por_app)
ax.legend_.remove()
plt.legend(loc="upper left", bbox_to_anchor=(1,1),title="id del app")
plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='20') # for legend title
figure = ax.get_figure()   
ax.set_xlabel("Número  de instalaciones", fontsize = 20)
ax.figure.set_size_inches(16, 12)
ax.set_ylabel("Dia", fontsize = 20)
ax.set_title('Cantidad de instalaciones según el día para el top 10 de las aplicaciones',fontsize = 25)
figure.savefig("Cantidad de instalaciones según el día para el top 10 de las aplicaciones.png")


plt.show()
```
