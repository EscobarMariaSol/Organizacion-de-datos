# Se busca visualizar la relación entre las application_id y los events_id

```python
appEvent = events[['event_id', 'application_id']]
appEvent.head()
```

```python
appEvent['count'] = 1
groupedEvApp = appEvent.groupby(['event_id','application_id']).sum().reset_index()
groupedEvApp.head()
```

```python
eventosConMasAplicaciones = groupedEvApp.nlargest(50, columns = 'count')
eventosConMasAplicaciones.head()
```

```python
eventAppPivoted = eventosConMasAplicaciones.pivot(index = 'event_id', columns = 'application_id', values = 'count')
eventAppPivoted.fillna(0, inplace = True)
eventAppPivoted.head()
```

```python
plt.figure(figsize = (20,9))
appVSeventGraph = sns.heatmap(data = eventAppPivoted, cmap = "Blues", linewidths = 0.2)
appVSeventGraph.set_title("Cantidad de id aplicaciones de cada id evento", fontsize = 20)
appVSeventGraph.set_xlabel("application_id", fontsize = 12)
appVSeventGraph.set_ylabel("event_id", fontsize = 12)
```

¿Cómo evolucionaron los eventos a través del tiempo?

```python
events['fecha'] = pd.to_datetime(events['date'], format = '%Y%m%d %H:%M:%S.%f').dt.date
```

```python
idVSfecha = events[['event_id','fecha']]
idVSfecha.head()
```

```python
idVSfecha['fecha'] = pd.to_datetime(idVSfecha['fecha']).dt.date
```

```python
idVSfecha.dtypes
```

```python
idVSfecha['count'] = 0
```

```python
pivoted = idVSfecha.pivot_table(index = 'event_id', columns = 'fecha', values = 'count', aggfunc = 'count')
pivoted = pivoted.fillna(0)
pivoted.head(10)
```

```python
plt.figure(figsize = (21,9))
idVSfechaGrapf = sns.heatmap(data = pivoted,linewidths = 0.1, cmap = "Blues")
idVSfechaGrapf.set_title("Cantidad de id aplicaciones por cada fecha", fontsize = 20)
idVSfechaGrapf.set_xlabel("fecha", fontsize = 12)
idVSfechaGrapf.set_ylabel("event_id", fontsize = 12)
```

¿Podemos mejorar esto? Pruebo filtrando los event_id que se hayan dado pocas veces a lo largo del tiempo

```python
pivotedFiltered = pivoted[pivoted.sum(axis = 1) > 10000]

pivotedFiltered.head()
```

```python
plt.figure(figsize = (20,10))
idVSfechaGrapf = sns.heatmap(data = pivotedFiltered, cmap = "Blues",linewidths = 0.1)
idVSfechaGrapf.set_title("Cantidad veces que se realiza la accion de un event_id por fecha", fontsize = 20)
idVSfechaGrapf.set_xlabel("Fecha", fontsize = 15)
idVSfechaGrapf.set_ylabel("Event_id", fontsize = 15)
```

Se ve que los eventos que más aparecen en cada fecha son los event_id 22, 31 y 161

A lo largo del día, ¿qué event_id aparece?

```python
events['hora'] = pd.to_datetime(events['date'], format = '%Y%m%d %H:%M:%S.%f').dt.hour
```

```python
events['fecha y hora'] = events['fecha'].astype(str)+' '+events['hora'].astype(str)
events.head()
```

```python
idVSdateHour = events[['event_id','fecha y hora']]
idVSdateHour['count'] = 0
idVSdateHour.head()
```

```python
pivotedDateAndHour = idVSdateHour.pivot_table(index = 'event_id', columns = 'fecha y hora', values = 'count', aggfunc = 'count')
pivotedDateAndHour = pivotedDateAndHour.fillna(0)
pivotedDateAndHour.head()
```

```python
pivotedDateAndHourFiltered = pivotedDateAndHour[pivotedDateAndHour.sum(axis = 1) > 50000]
pivotedDateAndHourFiltered.head()
```

```python
plt.figure(figsize = (20,10))
idVShourGrapf = sns.heatmap(data = pivotedDateAndHourFiltered, cmap = "Blues", linewidths = 0.1)
idVShourGrapf.set_title("Cantidad de eventos por ID de evento y por fecha con hora", fontsize = 20)
idVShourGrapf.set_xlabel("Fecha y Hora", fontsize = 15)
idVShourGrapf.set_ylabel("Event_id", fontsize = 15)
```

```python
pivotedDateAndHourFiltered['mean'] = pivotedDateAndHourFiltered.mean(axis=1)
pivotedDateAndHourFiltered.head()
```

```python
pivotedDateAndHourFiltered['mean'].nlargest(30).plot(kind = 'bar', figsize = (14,4))
```

Los tres eventos que mayor cantidad de frecuencia tienen son también los que tienen mayor promedio de aparición por cada hora de cada día

# Veo cuánto se realizan los eventos más frecuentes

```python
topEvents = events.loc[(events['event_id'] == 22) |\
                       (events['event_id'] == 31) |\
                       (events['event_id'] == 161) ][['event_id','hora']]
```

```python
topEvents.head()
```

```python
topEvents['count'] = 0
countEventByHour = topEvents.groupby(['event_id','hora']).count().reset_index()
countEventByHour.head()
```

```python
topEventsPivoted = topEvents.pivot_table(index = 'hora', columns = 'event_id', values = 'count', aggfunc = 'count')
topEventsPivoted.head()
```

```python
plt.figure(figsize = (20,9))
topEventsPivotedGraph = sns.heatmap(data = topEventsPivoted, cmap = "Blues", linewidths = 0.2)
topEventsPivotedGraph.set_title("Cantidad de veces que se dan los eventos mas comunes por hora", fontsize = 20)
topEventsPivotedGraph.set_xlabel("hora", fontsize = 12)
topEventsPivotedGraph.set_ylabel("event_id", fontsize = 12)
```
