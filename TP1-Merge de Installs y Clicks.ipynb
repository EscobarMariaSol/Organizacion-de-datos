# ¿Podemos mergear los dataframes installs y clicks?

Lo que se busca con el merge es verificar si un usuario tiende a instalar la aplicación luego de ver la publicidad, para eso se busca comparar las fechas de click y las fechas de instalación, y ver la diferencia entre ambas. Para analizar esto correctamente hay que asegurarse de estar analizando sobre exactamente el mismo dispositivo

```python
installs.dtypes
```

```python
clicks.dtypes
```

Asumiendo que la columna application_id del dataframe install y la columna advertiser_id del dataframe clicks refieren a la misma aplicación, se mergean los dataframes de installs y clicks. Luego, filtro el dataframe mergeado para poder asegurar el click y el install pertenecen al mismo dispositivo, sin embargo no se encuentra información suficiente para asegurarlo.

```python
merged = pd.merge(clicks, installs, left_on = "advertiser_id", right_on = "application_id", how = "inner")
merged = merged.loc[merged['country_code'] == merged['device_countrycode']]
merged.head()
```

```python
merged.size
```

```python
merged.dtypes
```

```python
merged['created_x'] = merged['created_x'].astype(str)
```

```python
merged['fecha_x'] = merged['created_x'].str[:10] + ' ' + merged['created_x'].str[11:23]
merged.head()
```

Ahora busco comparar las fechas de instalación y fechas de clicks para cada id de advertiser (asumiendo que es la equivalente a la id de la aplicación), para ello recupero la fecha de cada click e instalación con la mayor precisión posible.
Con esto se busca verificar si hay aplicaciones cuya instalación se realizó al poco tiempo de haber hecho click en la publicidad de la misma

```python
merged['fecha_x'] = pd.to_datetime(merged['fecha_x'])
```

```python
merged['fecha_y'] = merged['created_y'].str[:10] + ' ' + merged['created_y'].str[11:23]
merged.head()
```

```python
merged['fecha_y'] = pd.to_datetime(merged['fecha_y'])
```

```python
merged=merged.rename(columns = {'fecha_x':'fecha click', 'fecha_y':'fecha install'})
```

```python
mergedFechas = merged[['advertiser_id','fecha click','fecha install']]
```

Se crea una columna con la diferencia de tiempo entre la fecha de click y la fecha de instalación para cada advertiser_id, se entende que si una persona abre una publicidad y esta la convence de instalarla, la diferencia de tiempo entre instalación y click no debería ser muy grande.

```python
mergedFechas.head()
```

```python
mergedFechas['diferencia de tiempo'] = mergedFechas['fecha install'] - mergedFechas['fecha click']
mergedFechas.head()
```

Se entiende que la diferencia de tiempo no debería ser mayor a 20 minutos

```python
mergedFechasAppsInstaladas = mergedFechas\
        .loc[mergedFechas['diferencia de tiempo'] >= pd.Timedelta(0,'D')]\
        .loc[mergedFechas['diferencia de tiempo'] < pd.Timedelta(20,'m')]        
        #.loc[mergedFechas['diferencia de tiempo'] = pd.Timedelta(0,'h')]\
        #.loc[mergedFechas['diferencia de tiempo'] = pd.Timedelta(0,'m')]
```

```python
mergedFechasAppsInstaladas.sort_values(by = 'diferencia de tiempo', ascending = False)
```

```python
mergedFechasAppsInstaladas.size
```

No se consigue una cantidad suficiente de datos con este análisis para poder verificar si los usuarios realmente instalan las aplicaciones luego de haber dado click en su respectiva publicidad, además para los datos recibidos no puede asegurarse que el click y la instalación se hayan dado en el mismo dispositivo
